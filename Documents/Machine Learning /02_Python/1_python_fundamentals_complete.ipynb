# 1. CORE PYTHON SYNTAX 

## 1.1 Variables and Data Types

```python
# Variables (no declaration needed)
x = 10              # int
y = 3.14            # float
name = "Alice"      # str
is_valid = True     # bool

print(f"x = {x}, type: {type(x)}")
print(f"y = {y}, type: {type(y)}")
print(f"name = {name}, type: {type(name)}")
print(f"is_valid = {is_valid}, type: {type(is_valid)}")

# Multiple assignment
a, b, c = 1, 2, 3
print(f"\na = {a}, b = {b}, c = {c}")

# Same value to multiple variables
x = y = z = 0
print(f"x = {x}, y = {y}, z = {z}")
```

## 1.2 Type Conversion and Casting

```python
# String to int/float
s = "123"
num = int(s)
print(f"int('{s}') = {num}")

s = "3.14"
decimal = float(s)
print(f"float('{s}') = {decimal}")

# Int to string
n = 42
text = str(n)
print(f"str({n}) = '{text}'")

# Float to int (truncates decimal)
f = 3.99
i = int(f)
print(f"int({f}) = {i}")

# Boolean conversions
print(f"\nbool(0) = {bool(0)}")
print(f"bool(1) = {bool(1)}")
print(f"bool('') = {bool('')}")
print(f"bool('text') = {bool('text')}")
print(f"bool([]) = {bool([])}")
print(f"bool([1]) = {bool([1])}")
```

Incompatible Types:

Converting a string that does not represent a number (e.g., "abc") into a numeric type (e.g., int or float) will fail.
Example: int("abc") in Python raises a ValueError.


Lossy Conversions:

Converting a floating-point number to an integer may lose the fractional part, but it is technically possible. However, converting an integer to a floating-point number is always possible.

Complex Types:

Converting a complex number (e.g., 3 + 4j in Python) directly to an integer or float is not possible because the imaginary part cannot be represented in those types.


## 1.3 Operators

```python
# Arithmetic operators
a, b = 10, 3
print("Arithmetic Operators:")
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")     # True division (float)
print(f"{a} // {b} = {a // b}")   # Floor division (int) - Quotient without remainder
print(f"{a} % {b} = {a % b}")     # Modulo (remainder)
print(f"{a} ** {b} = {a ** b}")   # Exponentiation 

# Comparison operators
print("\nComparison Operators:")
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")
print(f"{a} > {b}: {a > b}")
print(f"{a} < {b}: {a < b}")
print(f"{a} >= {b}: {a >= b}")
print(f"{a} <= {b}: {a <= b}")

# Chaining comparisons
x = 5
print(f"\nChaining: 1 < {x} < 10: {1 < x < 10}")
```

## 1.4 Bitwise Operators (Important for LeetCode!)

```python
a, b = 5, 3  # Binary: 101 and 011

print("Bitwise Operators:")
print(f"{a} & {b} = {a & b}")   # AND: 001 = 1
print(f"{a} | {b} = {a | b}")   # OR:  101 = 5
print(f"{a} ^ {b} = {a ^ b}")   # XOR: 110 = 6
print(f"~{a} = {~a}")           # NOT: -6
print(f"{a} << 1 = {a << 1}")   # Left shift: 1010 = 10
print(f"{a} >> 1 = {a >> 1}")   # Right shift: 10 = 2

# Common bit tricks
n = 11
print(f"\n{n} is power of 2: {n & (n-1) == 0}")

n = 10
print(f"Check if bit 1 is set in {n}: {(n >> 1) & 1}")

# Toggle bit
n = 5  # 101
print(f"Toggle bit 0 in {n}: {n ^ (1 << 0)}")
```

# 2. CONTROL FLOW 

## 2.1 If-Elif-Else Statements

```python
# Basic if-else
x = 10
if x > 0:
    print(f"{x} is positive")
else:
    print(f"{x} is non-positive")

# If-elif-else
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
print(f"\nScore {score} -> Grade {grade}")

# Ternary operator (conditional expression)
x = 5
result = "even" if x % 2 == 0 else "odd"
print(f"\n{x} is {result}")

# Nested ternary
x = 0
result = "positive" if x > 0 else "negative" if x < 0 else "zero"
print(f"{x} is {result}")
```

## 2.2 For Loops

```python
# Basic for loop
print("Numbers 0-4:")
for i in range(5):
    print(i) # Print each number on a new line
print() # New line for separation

# Range with start and end
print("\nNumbers 5-9:")
for i in range(5, 10):
    print(i, end=' ') # Print on same line

# Range with step
print("\nEven numbers 0-8:")
for i in range(0, 10, 2):
    print(i, end=' ')
print()

# Enumerate (get index and value)
fruits = ["apple", "banana", "cherry"]
print("\nWith enumerate:")
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# Zip - iterate over multiple lists
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
print("\nWith zip:")
for name, age in zip(names, ages):
    print(f"  {name}, {age}")
```

## 2.3 While Loops

```python
# Basic while loop
count = 0
print("Count up to 5:")
while count < 5:
    print(count, end=' ')
    count += 1
print()

# While with break
print("\nFind first multiple of 7:")
n = 1
while True: # Infinite loop
    if n % 7 == 0: # Condition
        print(f"Found: {n}") 
        break # Exit loop
    n += 1 # Increment
    print(n, end=' ')

# While with continue
print("\nPrint odd numbers 1-9:")
n = 0
while n < 10: # Loop condition
    n += 1
    if n % 2 == 0: # Skip even numbers
        continue
    print(n, end=' ')
    
print()
```

## 2.4 List Comprehensions

```python
# Basic comprehension
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# With condition
evens = [x for x in range(10) if x % 2 == 0]
print("Evens:", evens)

# Conditional expression
parity = ['even' if x % 2 == 0 else 'odd' for x in range(5)]
print("Parity:", parity)

# Nested comprehension
matrix = [[i+j for j in range(3)] for i in range(3)]
print("\nMatrix:", matrix)
for row in matrix:
    print(row)
```

# 3. FUNCTIONS

## 3.1 Basic Functions

```python
# Basic function
def greet(name):
    """Greet someone by name."""
    return f"Hello, {name}!"

message = greet("Alice")
print(message)

# Function with multiple parameters
def add(a, b):
    """Add two numbers."""
    return a + b

result = add(5, 3)
print(f"5 + 3 = {result}")

# Function with default parameters
def power(base, exponent=2):
    """Calculate base raised to exponent."""
    return base ** exponent

print(f"\npower(3) = {power(3)}")
print(f"power(3, 3) = {power(3, 3)}")

# Multiple return values
def get_stats(numbers):
    """Return min, max, and sum of numbers."""
    return min(numbers), max(numbers), sum(numbers)

nums = [1, 2, 3, 4, 5]
min_val, max_val, total = get_stats(nums)
print(f"\nStats of {nums}: min={min_val}, max={max_val}, sum={total}")
```

## 3.2 *args and **kwargs

```python
# *args - variable number of positional arguments
def sum_all(*args):
    """Sum all arguments."""
    return sum(args)

print("sum_all(1, 2, 3):", sum_all(1, 2, 3))
print("sum_all(1, 2, 3, 4, 5):", sum_all(1, 2, 3, 4, 5))

# **kwargs - variable number of keyword arguments
def print_info(**kwargs):
    """Print all key-value pairs."""
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print("\nUser info:")
print_info(name="Alice", age=25, city="NYC")

# Combining positional, *args, and **kwargs
def flexible_func(required, *args, default="default", **kwargs):
    print(f"Required: {required}")
    print(f"Args: {args}")
    print(f"Default: {default}")
    print(f"Kwargs: {kwargs}")

print("\nFlexible function:")
flexible_func(1, 2, 3, default="custom", key1="value1")
```

## 3.3 Lambda Functions

```python
# Basic lambda
square = lambda x: x**2
print(f"square(5) = {square(5)}")

# Lambda with multiple arguments
add = lambda a, b: a + b
print(f"add(3, 4) = {add(3, 4)}")

# Lambda in sorted()
points = [(1, 2), (3, 1), (2, 3)]
print(f"\nOriginal: {points}")

# Sort by x-coordinate
sorted_x = sorted(points, key=lambda p: p[0])
print(f"Sorted by x: {sorted_x}")

# Sort by y-coordinate
sorted_y = sorted(points, key=lambda p: p[1])
print(f"Sorted by y: {sorted_y}")

# Sort by distance from origin
sorted_dist = sorted(points, key=lambda p: p[0]**2 + p[1]**2)
print(f"Sorted by distance: {sorted_dist}")

# Lambda with map()
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(f"\nDoubled: {doubled}")

# Lambda with filter()
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Evens: {evens}")
```

## 3.4 Recursion

```python
# Factorial
def factorial(n):
    """Calculate n! recursively."""
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print("Factorials:")
for i in range(6):
    print(f"  {i}! = {factorial(i)}")

# Fibonacci
def fibonacci(n):
    """Calculate nth Fibonacci number."""
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("\nFibonacci sequence:")
fib_nums = [fibonacci(i) for i in range(10)]
print(fib_nums)

# Fibonacci with memoization
def fib_memo(n, memo={}):
    """Fibonacci with memoization for efficiency."""
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

print(f"\nfib_memo(30) = {fib_memo(30)}")
```

## 4.1 Collections Module

### Counter - Counting Hashable Objects

```python
from collections import Counter

# Count elements in list
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
count = Counter(words)
print("Word count:", count)
print(f"'apple' appears {count['apple']} times")

# Count characters in string
text = "hello world"
char_count = Counter(text)
print(f"\nCharacter count in '{text}':")
print(char_count)

# Most common elements
print(f"\nMost common 2: {count.most_common(2)}")

# Counter arithmetic
c1 = Counter(['a', 'b', 'c', 'a'])
c2 = Counter(['a', 'b', 'd'])
print(f"\nc1: {c1}")
print(f"c2: {c2}")
print(f"c1 + c2: {c1 + c2}")
print(f"c1 - c2: {c1 - c2}")
print(f"c1 & c2 (intersection): {c1 & c2}")
print(f"c1 | c2 (union): {c1 | c2}")
```

### defaultdict - Dictionary with Default Values

```python
from collections import defaultdict

# defaultdict with int (default is 0)
word_count = defaultdict(int)
words = ['apple', 'banana', 'apple', 'cherry']
for word in words:
    word_count[word] += 1  # No need to check if key exists!

print("Word count:", dict(word_count))

# defaultdict with list
graph = defaultdict(list)
edges = [(1, 2), (1, 3), (2, 4), (3, 4)]
for src, dst in edges:
    graph[src].append(dst)

print("\nGraph:", dict(graph))

# defaultdict with set
char_indices = defaultdict(set)
text = "hello"
for i, char in enumerate(text):
    char_indices[char].add(i)

print("\nCharacter indices:", dict(char_indices))

# defaultdict with lambda
nested = defaultdict(lambda: defaultdict(int))
nested['user1']['score'] = 100
nested['user1']['level'] = 5
print("\nNested dict:", dict(nested))
```

### deque - Double-ended Queue

```python
from collections import deque

# Create deque
dq = deque([1, 2, 3])
print(f"Initial deque: {dq}")

# Add to right
dq.append(4)
print(f"After append(4): {dq}")

# Add to left
dq.appendleft(0)
print(f"After appendleft(0): {dq}")

# Remove from right
right = dq.pop()
print(f"After pop(): {dq}, popped: {right}")

# Remove from left
left = dq.popleft()
print(f"After popleft(): {dq}, popped: {left}")

# Extend
dq.extend([4, 5])
print(f"After extend([4, 5]): {dq}")

# Rotate
dq.rotate(1)  # Rotate right
print(f"After rotate(1): {dq}")

dq.rotate(-2)  # Rotate left
print(f"After rotate(-2): {dq}")

# Use as queue (FIFO)
queue = deque()
queue.append(1)  # enqueue
queue.append(2)
queue.append(3)
print(f"\nQueue: {queue}")
first = queue.popleft()  # dequeue
print(f"Dequeued: {first}, Remaining: {queue}")
```

### OrderedDict and namedtuple

```python
from collections import OrderedDict, namedtuple

# OrderedDict (Note: regular dict maintains order in Python 3.7+)
od = OrderedDict()
od['first'] = 1
od['second'] = 2
od['third'] = 3
print("OrderedDict:", od)

# Move to end
od.move_to_end('first')
print("After move_to_end('first'):", od)

# namedtuple - lightweight class
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(f"\nPoint: {p}")
print(f"x = {p.x}, y = {p.y}")
print(f"Also accessible by index: p[0] = {p[0]}, p[1] = {p[1]}")

# Unpacking
x, y = p
print(f"Unpacked: x = {x}, y = {y}")

# Use in data structures
points = [Point(1, 2), Point(3, 4), Point(5, 6)]
print(f"\nList of points: {points}")
```

## 4.2 Heapq Module - Priority Queue

### Min Heap Operations

```python
import heapq

# Create min heap
heap = []
heapq.heappush(heap, 5)
heapq.heappush(heap, 3)
heapq.heappush(heap, 7)
heapq.heappush(heap, 1)
print(f"Heap after pushes: {heap}")

# Pop smallest element
smallest = heapq.heappop(heap)
print(f"Popped smallest: {smallest}")
print(f"Heap after pop: {heap}")

# Peek at smallest (without removing)
print(f"Smallest element (peek): {heap[0]}")

# Heapify existing list
numbers = [5, 2, 8, 1, 9, 3]
heapq.heapify(numbers)
print(f"\nHeapified {[5, 2, 8, 1, 9, 3]}: {numbers}")

# Get n smallest elements
arr = [5, 2, 8, 1, 9, 3, 7]
smallest_3 = heapq.nsmallest(3, arr)
print(f"\n3 smallest in {arr}: {smallest_3}")

# Get n largest elements
largest_3 = heapq.nlargest(3, arr)
print(f"3 largest in {arr}: {largest_3}")
```

### Max Heap (Using Negation)

```python
import heapq

# Python only has min heap, so negate for max heap
max_heap = []
items = [5, 3, 7, 1, 9]

# Push negated values
for item in items:
    heapq.heappush(max_heap, -item)

print(f"Max heap (negated): {max_heap}")

# Pop and negate to get max
maximum = -heapq.heappop(max_heap)
print(f"Popped maximum: {maximum}")

# Peek at maximum
print(f"Current maximum (peek): {-max_heap[0]}")
```

### Heap with Custom Objects

```python
import heapq

# Using tuples (priority, data)
tasks = []
heapq.heappush(tasks, (3, 'low priority task'))
heapq.heappush(tasks, (1, 'high priority task'))
heapq.heappush(tasks, (2, 'medium priority task'))

print("Priority queue (tasks):")
while tasks:
    priority, task = heapq.heappop(tasks)
    print(f"  Priority {priority}: {task}")

# Top K frequent elements using heap
from collections import Counter

def top_k_frequent(nums, k):
    count = Counter(nums)
    # Use negative count for max heap
    return heapq.nlargest(k, count.keys(), key=count.get)

nums = [1, 1, 1, 2, 2, 3]
print(f"\nTop 2 frequent in {nums}: {top_k_frequent(nums, 2)}")
```

## 4.3 Itertools Module

### Infinite Iterators

```python
import itertools

# count - infinite counter
counter = itertools.count(start=10, step=2)
print("count(10, 2):")
for _ in range(5):
    print(next(counter), end=' ')
print()

# cycle - infinite cycle through iterable
cycler = itertools.cycle(['A', 'B', 'C'])
print("\ncycle(['A', 'B', 'C']):")
for _ in range(7):
    print(next(cycler), end=' ')
print()

# repeat - repeat element
repeater = itertools.repeat('X', 5)
print("\nrepeat('X', 5):")
print(list(repeater))
```

### Combinatoric Iterators

```python
import itertools

# Permutations - all orderings
items = ['A', 'B', 'C']
print("Permutations of ['A', 'B', 'C']:")
for perm in itertools.permutations(items):
    print(''.join(perm))

# Permutations of length 2
print("\nPermutations of length 2:")
for perm in itertools.permutations(items, 2):
    print(''.join(perm))

# Combinations - order doesn't matter
print("\nCombinations of ['A', 'B', 'C'] (length 2):")
for comb in itertools.combinations(items, 2):
    print(''.join(comb))

# Combinations with replacement
print("\nCombinations with replacement (length 2):")
for comb in itertools.combinations_with_replacement(['A', 'B'], 2):
    print(''.join(comb))

# Product - Cartesian product
print("\nProduct of ['A', 'B'] and [1, 2]:")
for item in itertools.product(['A', 'B'], [1, 2]):
    print(item)
```

### Useful Itertools Functions

```python
import itertools

# chain - combine iterables
list1 = [1, 2, 3]
list2 = [4, 5, 6]
chained = itertools.chain(list1, list2)
print("chain([1,2,3], [4,5,6]):", list(chained))

# islice - slice an iterator
numbers = range(100)
sliced = itertools.islice(numbers, 5, 15, 2)
print("\nislice(range(100), 5, 15, 2):", list(sliced))

# groupby - group consecutive equal elements
data = [1, 1, 2, 2, 2, 3, 1, 1]
print("\ngroupby([1,1,2,2,2,3,1,1]):")
for key, group in itertools.groupby(data):
    print(f"  {key}: {list(group)}")

# accumulate - running totals
numbers = [1, 2, 3, 4, 5]
cumsum = itertools.accumulate(numbers)
print(f"\naccumulate({numbers}):", list(cumsum))

# accumulate with operator
import operator
cumproduct = itertools.accumulate(numbers, operator.mul)
print(f"accumulate with multiply:", list(cumproduct))

# zip_longest - zip with fillvalue
list1 = [1, 2, 3]
list2 = ['a', 'b']
zipped = itertools.zip_longest(list1, list2, fillvalue='X')
print(f"\nzip_longest({list1}, {list2}):", list(zipped))
```

# 5. PYTHON IDIOMS & BEST PRACTICES üí°

## 5.1 Pythonic Code

```python
# Swapping variables
a, b = 5, 10
print(f"Before: a={a}, b={b}")
a, b = b, a  # ‚úÖ Pythonic
print(f"After swap: a={a}, b={b}")

# Multiple assignment
x, y, z = 1, 2, 3
print(f"\nMultiple assignment: x={x}, y={y}, z={z}")

# Extended unpacking
first, *middle, last = [1, 2, 3, 4, 5]
print(f"\nExtended unpacking:")
print(f"  first={first}, middle={middle}, last={last}")

# Underscore for unused variables
for _ in range(3):
    print("Hello", end=' ')
print()

# Chaining comparisons
x = 5
if 1 < x < 10:  # ‚úÖ Instead of: if x > 1 and x < 10
    print(f"\n{x} is between 1 and 10")

# Using in for membership
vowels = 'aeiou'
letter = 'a'
if letter in vowels:  # ‚úÖ Pythonic
    print(f"{letter} is a vowel")
```

## 5.2 List/Dict/Set Comprehensions

```python
# List comprehension
# ‚ùå Not Pythonic
squares = []
for x in range(10):
    squares.append(x**2)

# ‚úÖ Pythonic
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# Dict comprehension
# ‚úÖ Pythonic
square_dict = {x: x**2 for x in range(5)}
print("\nSquare dict:", square_dict)

# Set comprehension
# ‚úÖ Pythonic
unique_lengths = {len(word) for word in ['apple', 'banana', 'kiwi', 'pear']}
print("Unique lengths:", unique_lengths)

# Generator expression (memory efficient)
sum_of_squares = sum(x**2 for x in range(1000000))  # Doesn't create list
print(f"\nSum of first 1M squares: {sum_of_squares}")
```

## 5.3 Enumerate and Zip

```python
fruits = ['apple', 'banana', 'cherry']

# ‚ùå Not Pythonic
print("Not Pythonic:")
for i in range(len(fruits)):
    print(f"  {i}: {fruits[i]}")

# ‚úÖ Pythonic - enumerate
print("\nPythonic with enumerate:")
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# Enumerate with custom start
print("\nWith start=1:")
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}: {fruit}")

# Zip for parallel iteration
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
print("\nZip example:")
for name, age in zip(names, ages):
    print(f"  {name} is {age} years old")

# Creating dict from two lists
person_dict = dict(zip(names, ages))
print(f"\nDict from zip: {person_dict}")
```

## 5.4 Any and All

```python
# any() - returns True if any element is True
numbers = [0, 0, 5, 0]
print(f"any({numbers}): {any(numbers)}")

# Check if any even
nums = [1, 3, 5, 7, 8]
has_even = any(x % 2 == 0 for x in nums)
print(f"\nHas even number: {has_even}")

# all() - returns True if all elements are True
numbers = [1, 2, 3, 4]
print(f"\nall({numbers}): {all(numbers)}")

numbers = [1, 2, 0, 4]
print(f"all({numbers}): {all(numbers)}")

# Check if all positive
nums = [1, 2, 3, 4]
all_positive = all(x > 0 for x in nums)
print(f"\nAll positive: {all_positive}")

# Practical example
def is_valid_triangle(a, b, c):
    sides = [a, b, c]
    return all(x > 0 for x in sides) and \
           a + b > c and b + c > a and a + c > b

print(f"\nis_valid_triangle(3, 4, 5): {is_valid_triangle(3, 4, 5)}")
print(f"is_valid_triangle(1, 2, 10): {is_valid_triangle(1, 2, 10)}")
```

## 5.7 Exception Handling Best Practices

```python
# ‚úÖ Specific exceptions
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Error: Division by zero")
        return None
    except TypeError:
        print("Error: Invalid types")
        return None

print("divide(10, 2):", divide(10, 2))
print("divide(10, 0):", divide(10, 0))

# EAFP: Easier to Ask for Forgiveness than Permission
d = {'a': 1, 'b': 2}

# ‚ùå LBYL (Look Before You Leap)
if 'c' in d:
    value = d['c']
else:
    value = 0

# ‚úÖ EAFP (Pythonic)
try:
    value = d['c']
except KeyError:
    value = 0

# Or simply use get()
value = d.get('c', 0)  # ‚úÖ Even better
print(f"\nValue of 'c': {value}")
```

## 5.8 List Operations Best Practices

```python
# Avoid modifying list while iterating
# ‚ùå Bad
numbers = [1, 2, 3, 4, 5, 6]
# for i, num in enumerate(numbers):
#     if num % 2 == 0:
#         numbers.pop(i)  # Can cause issues

# ‚úÖ Good - create new list
numbers = [1, 2, 3, 4, 5, 6]
odds = [num for num in numbers if num % 2 != 0]
print(f"Odd numbers: {odds}")

# Reversing
numbers = [1, 2, 3, 4, 5]

# ‚úÖ Create new reversed list
reversed_nums = numbers[::-1]
print(f"\nReversed (new list): {reversed_nums}")
print(f"Original: {numbers}")

# ‚úÖ Reverse in place
numbers.reverse()
print(f"Reversed in place: {numbers}")

# Flattening nested lists
nested = [[1, 2], [3, 4], [5, 6]]

# ‚úÖ List comprehension
flat = [item for sublist in nested for item in sublist]
print(f"\nFlattened: {flat}")

# ‚úÖ Or use itertools.chain
import itertools
flat = list(itertools.chain(*nested))
print(f"Flattened (chain): {flat}")
```

## Problem 1: Two Sum

```python
def two_sum(nums, target):
    """
    Find two numbers that add up to target.
    Returns indices.
    """
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Test
nums = [2, 7, 11, 15]
target = 9
result = two_sum(nums, target)
print(f"Input: {nums}, target: {target}")
print(f"Output: {result}")
print(f"Verification: {nums[result[0]]} + {nums[result[1]]} = {target}")
```

## Problem 2: Top K Frequent Elements

```python
from collections import Counter
import heapq

def top_k_frequent(nums, k):
    """
    Find k most frequent elements.
    """
    # Method 1: Using Counter
    count = Counter(nums)
    return [num for num, freq in count.most_common(k)]

def top_k_frequent_heap(nums, k):
    """
    Using heap for O(n log k) solution.
    """
    count = Counter(nums)
    return heapq.nlargest(k, count.keys(), key=count.get)

# Test
nums = [1, 1, 1, 2, 2, 3]
k = 2
print(f"Input: {nums}, k: {k}")
print(f"Top {k} frequent (Counter): {top_k_frequent(nums, k)}")
print(f"Top {k} frequent (Heap): {top_k_frequent_heap(nums, k)}")
```

## Problem 3: Valid Anagram

```python
from collections import Counter

def is_anagram(s, t):
    """
    Check if t is anagram of s.
    """
    # Method 1: Counter
    return Counter(s) == Counter(t)

def is_anagram_sort(s, t):
    """Method 2: Sorting."""
    return sorted(s) == sorted(t)

def is_anagram_dict(s, t):
    """Method 3: Manual counting."""
    if len(s) != len(t):
        return False
    
    count = {}
    for char in s:
        count[char] = count.get(char, 0) + 1
    
    for char in t:
        if char not in count or count[char] == 0:
            return False
        count[char] -= 1
    
    return True

# Test
test_cases = [
    ("anagram", "nagaram", True),
    ("rat", "car", False),
    ("listen", "silent", True)
]

for s, t, expected in test_cases:
    result = is_anagram(s, t)
    status = "‚úÖ" if result == expected else "‚ùå"
    print(f"{status} '{s}' and '{t}': {result}")
```

## Problem 4: Group Anagrams

```python
from collections import defaultdict

def group_anagrams(words):
    """
    Group words that are anagrams.
    """
    groups = defaultdict(list)
    
    for word in words:
        # Use sorted characters as key
        key = ''.join(sorted(word))
        groups[key].append(word)
    
    return list(groups.values())

# Test
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(f"Input: {words}")
print("\nGrouped anagrams:")
for group in group_anagrams(words):
    print(f"  {group}")
```

## Problem 5: Kth Largest Element

```python
import heapq

def find_kth_largest(nums, k):
    """
    Find kth largest element in array.
    """
    # Method 1: Using heapq.nlargest
    return heapq.nlargest(k, nums)[-1]

def find_kth_largest_heap(nums, k):
    """
    Using min heap of size k.
    """
    heap = nums[:k]
    heapq.heapify(heap)
    
    for num in nums[k:]:
        if num > heap[0]:
            heapq.heapreplace(heap, num)
    
    return heap[0]

# Test
nums = [3, 2, 1, 5, 6, 4]
k = 2
print(f"Array: {nums}")
print(f"{k}nd largest: {find_kth_largest(nums, k)}")
print(f"{k}nd largest (heap): {find_kth_largest_heap(nums, k)}")
```
